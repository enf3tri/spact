# ***Extracting Frames from Video***


```python
import cv2

video_path = ''
vidcap = cv2.VideoCapture(video_path)
success,image = vidcap.read()
count = 0

while success:
  cv2.imwrite("frame%d.jpg" % count, image)     
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 20 #Reducing frames

```
